```python
import torch
import torch.nn.functional as F
```

```python
def TracedModelFactory(file_name, traced_model):
    traced_model.save(file_name)
    traced_model = torch.jit.load(file_name)
    print("filename : ", file_name)
    print(traced_model.graph)
```

```python
from torchvision import models as tv_models
```

```python
img = torch.randn((1, 3, 224,224))
model = tv_models.AlexNet()
model.eval()
model = model.cpu()
traced_model = torch.jit.trace(model, (img))

TracedModelFactory("alexnet.pth", traced_model)
```
